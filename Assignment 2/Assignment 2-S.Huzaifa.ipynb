### Step 1: Collect data to create visuals and speak with numbers

>>Many a statistic is false on its face. It gets by only because the magic of numbers brings about a suspension of common sense.
-Darrell Huff

That's right! Audience is more likely to believe a message if the message is supported with numbers because then it makes the message seem more researched and official. 

```python
import pandas as pd #import pandas library analyze data  
import matplotlib.pyplot as plt

#Importing world climate data and view data info (datasource: Berkeley Earth URL: http://berkeleyearth.org/data/)
world_climate = pd.read_csv("data/GlobalLandTemperaturesByCity.csv",  parse_dates=["dt"])
world_climate.info()

#Create a column "Year". This will help us filter the data and get average temperatures per year. 

world_climate["Year"] = world_climate["dt"].dt.year
```

```python
plt.style.use("seaborn-deep")
plt.figure(figsize=(10,8), dpi=150)
#Grouping by the year, calculating average world temperature each year.
world_climate.groupby("Year")["AverageTemperature"].mean().tail(50).plot(kind="line"
                                                                         ,title="Average Temperature in the past 50 Years"
                                                                         ,ylabel="Temperature(Celsius)"
                                                                         ,linewidth=10
                                                                         ,color='lightblue'
                                                                         ,label="World"
                                                                         ) 

#To make it more appealing to Canadians, lets look into Canada's climate 
can_climate = world_climate[world_climate["Country"] == 'Canada']
  

#grouping by year to get the yearly Candian average temperatures
can_climate.groupby("Year")["AverageTemperature"].mean().tail(50).plot(kind="line"
                                                                  ,ylabel="Temperature(Celsius)"
                                                                  ,linewidth=10
                                                                      ,color='#5dc9b6'
                                                                      ,label="Canada")

#Setting a ridiculous y-axis range to make the temperature change visually insignificant.
plt.ylim(-50,100) 
plt.legend(borderpad=2, labelspacing=2)
plt.show()
```
